```python
import requests, bs4
from bs4 import BeautifulSoup as bs
import pandas as pd
import pickle
import time
```

```python
with open('cars_url_list.pickle', 'rb') as handle:
    cars_url_list = pickle.load(handle)
```

```python
cars_url_list
```

```python
def get_zip(soup):
    try:
        zip=soup.find(class_="dealer-address").text.split(' ')[-1]
    except:
        zip=None
    return zip
        
    
```

```python
zip_list=[]
```

```python
i=0
for url in cars_url_list:
    response = requests.get(url)
    page = response.text
    soup = bs(page)
    zip_code=get_zip(soup)
    zip_list.append(zip_code)
    i+=1
    if i%50==0:
        print(i,zip_list[i-1])
```

```python
len(zip_list)
```

```python
with open('zip_list.pickle', 'wb') as handle:
    pickle.dump(zip_list, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
